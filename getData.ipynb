```python
from pyscopus import Scopus
import json
import os
import numpy as np
import requests
    
## Load configuration
filename = os.getcwd() + "/elsapy_/config.json"
with open(filename) as f:
    config = json.load(f)

scopus = Scopus(config['apikey'])
```

```python
search_df = scopus.search("KEY(science)", count=5000)
```

```python
import pickle

# with open('/Users/alexhuang/Google Drive/Intro-ML-Scientific-Paper-NLP-Project/search_df.pkl', 'rb') as infile:  # Overwrites any existing file.
#     search_df = pickle.load(infile)
    
def get_full_text(s):
    try:
        return scopus.retrieve_full_text('https://api.elsevier.com/content/article/scopus_id/' + str(s))
    except:
        #print("Could not find")
        return np.nan
        
def get_abstract(s):
    try:
        return scopus.retrieve_abstract(s)['abstract']
    except:
        #print("Could not find abstract.")
        return np.nan
    
search_df['abstract'] = search_df['scopus_id'].apply(lambda s: get_abstract(s))
search_df['full_text'] = search_df['scopus_id'].apply(lambda s: get_full_text(s))

with open('/Users/alexhuang/Google Drive/Intro-ML-Scientific-Paper-NLP-Project/search_df.pkl', 'wb') as output:  # Overwrites any existing file.
    pickle.dump(search_df,output,pickle.HIGHEST_PROTOCOL)
```
